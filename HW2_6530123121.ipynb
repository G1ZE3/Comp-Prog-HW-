```python
# HW02 (ไม่ลบหรือแก้ไขบรรทัดนี้ หรือเพิ่มอะไรก่อนบรรทัดนี้ โดยเด็ดขาด)

def get_sum(data):
 d = list(data)
 d.append('q')
 get = []
 person = [0]*2
 person[0] = d[0]
 score = []
 for i in range(0,len(d),2):
     if d[i] == person[0]:
         score.append(int(d[i+1]))
     elif d[i] == 'q':
       if len(score)>3:
         score.remove(min(score))
       person[1] = sum(score)
       get += person
     else:
       if len(score)>3:
         score.remove(min(score))
       person[1] = sum(score)
       get += person
       person[0] = d[i]
       score = []
       score.append(int(d[i+1]))
 return get
```
